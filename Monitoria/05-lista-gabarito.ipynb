**Questão 3)** Faça uma função que receba uma frase (string) e retorne quantos caracteres são espaços em brancos.


```python
def espacosBrancos(frase):
    num = 0

    for item in frase:
        if item == ' ':
            num += 1
            
    return num
```

**Questão 4)** Escreva um programa que leia a idade e o primeiro nome de $n$ pessoas. Seu programa
deve terminar quando uma idade negativa for digitada. Ao terminar, seu programa deve
escrever o nome da pessoa mais jovem e da mais velha.

**Exemplos:**
<pre>
Entrada:
   João
   45
   Maria
   15
   Hugo
   22
Saída:
   Maria
   João

Entrada:
   Alex
   5
   José
   13
   Jeremias
   6
   Márcia
   15
Saída:
   Alex
   Márcia
   
Entrada:
    Ronaldo
    24
Saída:
    Ronaldo
    Ronaldo
</pre>

```python
def nome_idade():
    nome = input('Digite o nome: ')
    idade = int(input('Digite a idade: '))
    
    return nome, idade

maiorNome, maiorIdade = nome_idade()

menorNome = maiorNome
menorIdade = maiorIdade

while True:
    nome, idade = nome_idade()
    
    if idade < 0:
        break
        
    if idade > maiorIdade:
        maiorIdade = idade
        maiorNome = nome
        
    elif idade < menorIdade:
        menorIdade = idade
        menorNome = nome
        
print(menorNome)
print(maiorNome)
```

**Questão 5)** Escreva uma função que, dada uma string, retorne True se é um palı́ndromo e False, caso contrário. Lem-
brando que um palı́ndromo é uma palavra que tenha a propriedade de poder ser lida tanto da direita para a esquerda como da esquerda para a direita. Sua função deve ignorar símbolos, pontuação e espaços. 

Utilize a função para implementar um programa capaz de executar os exemplos abaixo corretamente.

**Exemplos:**
<pre>
Entrada:
   ovo
Saída:
   Palíndromo!

Entrada:
   Batata
Saída:
   Não é palíndromo!
   
Entrada:
    Anotaram a data da maratona
Saída:
    Palíndromo!
    
Entrada:
    arara
Saída:
    Palíndromo!    
</pre>

```python
string = input('Digite um string: ')

j = len(string) - 1
lim = int(len(string) / 2)

palindromo = True

for i in range(0, lim):
    if string[i] is not string[j]:
        palindromo = False
        break
        
    j -= 1

if palindromo:
    print('Palíndromo!')
    
else:
    print('Não é palíndromo!') 
```

**Questão 8)** Faça uma função que receba uma sequência de digitos (em formato string) e encontre o conjunto de 3 dı́gitos consecutivos que gere o maior produto. A função deve retornar uma string com os três digitos consecutivos encontrados.
A função deve retornar a primeira sequência encontrada, no caso de mais de uma válida.

**Exemplos:**
<pre>
Entrada:
   73167176531330624919225119674426574742355349198934
Saída:
   989

Entrada:
   00100010001010000000000000011000000000100000000000
Saída:
   001
   
Entrada:
    1253701707070231231231237560123701231234441111231
Saída:
    756
</pre>

```python
def digitos(string):

    maiorVal = -1
    maiorSeq = None

    for i in range (0, len(string) - 2):
        seq = string[i:i+3]

        num1 = int(string[i])
        num2 = int(string[i+1])
        num3 = int(string[i+2])

        val = num1 * num2 * num3

        if val > maiorVal:
            maiorVal = val
            maiorSeq = seq

    return maiorSeq
```

**Questão 9)** Escreva um programa que leia uma frase e duas palavras do teclado. O programa deve imprimir a frase substibuindo todas as ocorrências da primeira palavra pela segunda. O programa também deve imprimir uma mensagem indicando quantas substituições foram realizadas. Apenas tratar casos de pontuação com ", . : ?".

**Exemplos:**
<pre>
Entrada:
   a ligeira raposa marrom saltou sobre o cachorro cansado
   cachorro
   gato
Saída:
   a ligeira raposa marrom saltou sobre o gato cansado
   Substituições realizadas: 1

Entrada:
   um tigre, dois tigres, tres tigres tristes
   tigres
   marrecos
Saída:
   um tigre, dois marrecos, tres marrecos tristes
   Substituições realizadas: 2
   
Entrada:
   Se a liga me ligasse, eu ligava a liga, mas como a liga não me liga, eu não ligo a liga
   a
   o
Saída:
   Se o liga me ligasse, eu ligava o liga, mas como o liga não me liga, eu não ligo o liga
   Substituições realizadas: 4
</pre>

```python
frase = input('Digite a frase: ')
substituida = input('Digite a palavra que vai ser substituída: ')
substituidora = input('Digite a palavra que vai substituir: ')

substituida = ' ' + substituida
substituidora = ' ' + substituidora

num = frase.count(substituida)
novaFrase = frase.replace(substituida, substituidora)

print(novaFrase)
print('Substituições realizadas:', num)
```
