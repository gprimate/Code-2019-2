**Questão 1)** Escreva uma função chamada *escalas* que recebe um arquivo como parâmetro: a primeira linha do arquivo representa o *inicio*, a segunda o *fim* e terceira o *passo*, referentes a um intervalo de temperatura em graus Celsius. Sua função deve imprimir *n* linhas contendo  uma tabela de conversão de graus Celsius para Fahrenheit e Kelvin **nessa ordem**. Escreva também o código principal (fora da função) onde os valores *inicio*, *fim* e *passo* serão lidos do usuário.

**Nota:** A conversão de Celsius para Fahrenheit é dada por: $TF = TC *\frac{9}{5} + 32$ <br>
A conversão de Celsius para Kelvin é dada por: $TK = TC + 273$

**Exemplo**
<pre>
Formato do arquivo:
    50
    100
    10
Saida:
    50 122 323
    60 140 333
    70 158 343
    80 176 353
    90 194 363
    100 212 373
</pre>

```python
def escalas(arquivo):
    
    linhas = arquivo.readlines()
    numeros = [int(num.strip()) for num in linhas]
    
    for i in range(numeros[0], numeros[1] + 1, numeros[2]):
        print(i, int(i*9/5 + 32), i + 273)
```

**Questão 2)** Escreva a função *mdc4(arquivo)* que retorna o máximo divisor comum entre 4 números presentes no arquivo, todos na mesma linha separados por espaço.

Dica: O mdc(a,b,c) = mdc(a, mdc(b,c) )


**Exemplo**
<pre>
Formato do arquivo:
    156 182 429 26
Saida:
    13
    
Formato do arquivo:
    6 12 15 18
Saida:
    3
</pre>


```python
def mdc4(arquivo):
    num = []
    
    for linha in arquivo:
        for n in linha.split():
            num.append(int(n))
            
    def mdc(x,y):
        resto = None
        
        while resto is not 0:
            resto = x % y
            x  = y
            y  = resto
            
        return x
    
    return mdc(mdc(num[0],num[1]), mdc(num[2],num[3]))
```

**Questão 3)** Escreva uma função valores(arquivo) que recebe um arquivo como parâmetro que possui n números (1 por linha) e que calcule:
<ol>
<li>A soma de todos os números</li>
<li>O maior número digitado</li>
<li>O menor número digitado</li>
<li>A média dos números pares</li>
<li>A quantidade de primos na sequência</li>
</ol>

**Formato do arquivo (exemplo):** 
<pre>
    25
    2
    47
    5
    8
    90
<pre>

```python
def valores(arquivo):
    linhas = arquivo.readlines()
    numeros = [int(num.strip()) for num in linhas]
    
    print('A soma é:', sum(numeros))
    print('O maior número é:', max(numeros))
    print('O menor número é:', min(numeros))
    
    n = 0
    total = 0
    nprimos = 0

    for i in numeros:
        primo = True 

        if i % 2 == 0:
            total += i
            n += 1

        if i <= 1:
            primo = False

        for j in range(2,i):
            if i % j == 0:
                primo = False
                break

        if primo == True:
            nprimos += 1

    print('A média dos pares é:', total/n)
    print('O número de primos é:', nprimos)
```

**Questão 4)** Faça uma função que retorne o número de 1’s que aparecem em um string recebido por um arquvo.

**Exemplos:**
<pre>
Formato do arquivo:
   0011001
Saída:
   3

Formato do arquivo:
   11
Saída:
   2
</pre>

```python
def num1s (arquivo):
    linhas = arquivo.readlines()
    num = 0
    
    for item in linhas[0]:
        if item == '1':
            num += 1
    return num
```

**Questão 5)** Faça uma função que receba um arquivo com uma frase e retorne quantos caracteres são espaços em brancos.

```python
def espacosBrancos(arquivo):
    num = 0
    
    for linha in arquivo:
        for letra in linha:
            if letra == ' ':
                num += 1
            
    return num
```

**Questão 6)** Escreva um função que leia um arquvo que contém a idade e o primeiro nome de $n$ pessoas. Sua função deve
escrever o nome da pessoa mais jovem e da mais velha.

**Exemplos:**
<pre>
Formato arquivo:
   João
   45
   Maria
   15
   Hugo
   22
Saída:
   Maria
   João

Formato arquivo:
   Alex
   5
   José
   13
   Jeremias
   6
   Márcia
   15
Saída:
   Alex
   Márcia
   
Formato arquivo:
    Ronaldo
    24
Saída:
    Ronaldo
    Ronaldo
</pre>

```python
def maior_menor(arquivo):
    linhas = arquivo.readlines()
    itens = [item.strip() for item in linhas]

    idades = []
    nomes = []
    i_maior = 0
    i_menor = 0

    for item in itens:
        if item.isnumeric():
            idades.append(int(item))
        else:
            nomes.append(item)

            
    for i, idade in enumerate(idades):
        if idade > idades[i_maior]:
            i_maior = i

        elif idade < idades[i_menor]:
            i_menor = i
            
    print(nomes[i_menor])
    print(nomes[i_maior])
```

**Questão 10)** Escreva uma função que recebe um arquivo com um nome de pessoa em cada linha e retorne uma lista com os nomes em ordem alfabética.

**Exemplos:**
<pre>
Formato arquivo:
   João
   Maria
   Hugo
   Ziad
   Samuel
Saída:
   Hugo
   João
   Maria
   Samuel
   Ziad
   
Formato arquivo:
   José Maria
   José Silva
   José Carlos
Saída:
   José Carlos
   José Maria
   José Silva
   
Formato arquivo:
    Romário
    Rivaldo
    Ronaldo
Saída:
    Rivaldo
    Romário
    Ronaldo
</pre>

```python
def alfabetica(arquivo):
    linhas = arquivo.readlines()
    nomes = [nome.strip() for nome in linhas]
    
    nomes.sort()
    
    return nomes
```
